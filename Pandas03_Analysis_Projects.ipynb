# 01 Clean Dataset of First_Health_Camp_Attended.csv


```python
import pandas as pd
# Load the employee data from the CSV file from the provided GitHub URL
df = pd.read_csv('https://raw.githubusercontent.com/itsluckysharma01/Datasets/refs/heads/main/Health_Care_Dataset/First_Health_Camp_Attended.csv')
print("Data loaded successfully!")

print("First 5 rows of the dataset:")
print(df.head())
# Display the shape of the DataFrame
print("\nShape of the DataFrame:")
print(df.shape)
# Display the columns of the DataFrame
print("\nColumns in the DataFrame:")
print(df.columns)

print(df.info())
print(df.describe())

#Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

#handle NnaN values
#Drop rows with any NaN values
df_cleaned = df.dropna(inplace=True)
print("\nData after dropping rows with NaN values:")
print(df_cleaned)
print(df.isnull().sum())

print("SUCCESSFULLY LOADED THE DATASET AND HANDLED MISSING VALUES!")

#Export the cleaned DataFrame to a new CSV file
# df.to_csv('F:\PROGRAMMING\MY PROJECTS\Datasets\Health_Care_Dataset\Cleaned_Data\cleaned_health_camp_Attended_data.csv', index=False)
# print("Cleaned data exported  successfully!")
```


# 02 Clean Dataset of Health_Camp_Detail.csv


```python
import pandas as pd
# Load the employee data from the CSV file from the provided GitHub URL
df = pd.read_csv('https://raw.githubusercontent.com/itsluckysharma01/Datasets/refs/heads/main/Health_Care_Dataset/Health_Camp_Detail.csv')
print("Data loaded successfully!")

print("First 5 rows of the dataset:")
print(df.head())
print("\nShape of the DataFrame:")
print(df.shape)
print("\nColumns in the DataFrame:")
print(df.columns)
print(df.info())
print(df.describe())

#Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
print("\nData is already clean, no NaN values found.")

#check for duplicate rows
print("\nChecking for duplicate rows:")
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

print("\nData after checking for duplicates:")
if duplicates > 0:
    df = df.drop_duplicates()
    print("Duplicates removed.")
else:
    print("No duplicates found.")

#Export the cleaned DataFrame to a new CSV file
# df.to_csv('F:\PROGRAMMING\MY PROJECTS\Datasets\Health_Care_Dataset\Cleaned_Data\cleaned_health_camp_detail.csv', index=False)
# print("Cleaned data exported  successfully!")
```
