The **Seaborn** library in Python is a powerful and user-friendly data visualization library based on **Matplotlib**. It is widely used for creating **statistical graphics** and is especially popular in data science and analytics for making complex plots with minimal code.

---

## üåä What is Seaborn?

**Seaborn** is built on top of **Matplotlib** and integrates closely with **Pandas** data structures. It provides a high-level interface for drawing attractive and informative **statistical graphics**.

```bash
pip install seaborn
```

---

## üì¶ Key Features of Seaborn

| Feature                          | Description                                      |
| -------------------------------- | ------------------------------------------------ |
| High-level interface             | Simplifies the process of creating complex plots |
| Built-in themes                  | Makes plots look visually appealing              |
| Works well with Pandas           | Accepts DataFrames directly                      |
| Automatic statistical estimation | Supports confidence intervals, trend lines, etc. |
| Support for categorical data     | Bar plots, box plots, violin plots, etc.         |
| Color palettes                   | Beautiful color schemes for consistent styling   |

---

## üõ†Ô∏è Commonly Used Seaborn Functions

| Function                        | Description                                         |
| ------------------------------- | --------------------------------------------------- |
| `sns.lineplot()`                | Line plot for continuous data                       |
| `sns.barplot()`                 | Bar chart with optional statistical estimation      |
| `sns.countplot()`               | Bar chart for categorical counts                    |
| `sns.boxplot()`                 | Box-and-whisker plot                                |
| `sns.violinplot()`              | Combines boxplot with KDE                           |
| `sns.scatterplot()`             | Scatter plot for bivariate data                     |
| `sns.heatmap()`                 | Heatmap for matrix-like data                        |
| `sns.pairplot()`                | Pairwise relationships in a dataset                 |
| `sns.distplot()` *(deprecated)* | Distribution plot (use `histplot` or `kdeplot` now) |
| `sns.histplot()`                | Histogram with optional KDE                         |
| `sns.kdeplot()`                 | Kernel Density Estimate plot                        |
| `sns.catplot()`                 | Generic function for categorical plots              |
| `sns.jointplot()`               | Bivariate scatter plots with histograms             |

---

## üé® Styling and Themes

You can customize plots using themes and color palettes:

```python
import seaborn as sns

sns.set_style("whitegrid")   # Options: white, dark, ticks, whitegrid, darkgrid
sns.set_palette("pastel")    # Or "deep", "muted", "bright", etc.
```

## üìä Example: Simple Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load sample dataset
df = sns.load_dataset('tips')

# Create a bar plot
sns.barplot(x='day', y='total_bill', data=df)

plt.title("Average Bill per Day")
plt.show()
```

## üîÅ Seaborn vs Matplotlib

| Feature        | Seaborn                          | Matplotlib                 |
| -------------- | -------------------------------- | -------------------------- |
| Syntax         | High-level, simple               | Low-level, detailed        |
| Styling        | Beautiful by default             | Needs manual configuration |
| Data Structure | Works natively with Pandas       | Needs arrays or lists      |
| Plot Types     | Specializes in statistical plots | General-purpose            |

# Matrix Plots
***
# Heatmap


````python
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap of Flights")
plt.show()
````

# Clustermap

````python
sns.clustermap(flights_pivot, cmap="coolwarm")
plt.show()
````
