Diamond Pattern

```python
# Print diamond pattern
n = int(input("Enter number of rows (half of diamond): "))

for i in range(n):
    print(" " * (n - i - 1) + "* " * (i + 1))

for i in range(n - 2, -1, -1):
    print(" " * (n - i - 1) + "* " * (i + 1))
```

Hollow Square Pattern

```python
# Print hollow square pattern
n = int(input("Enter size of square: "))
for i in range(n):
    if i == 0 or i == n - 1:
        print("* " * n)
    else:
        print("* " + "  " * (n - 2) + "*")
```

Hexagonal Star Pattern

```python
# Print hexagonal pattern
n = int(input("Enter size of hexagon (number of rows for half): "))

# Upper part
for i in range(n):
    print(" " * (n - i - 1) + "* " * (n + i))

# Lower part
for i in range(n - 2, -1, -1):
    print(" " * (n - i - 1) + "* " * (n + i))
```

Circular star Pattern

```python
import math

radius = int(input("Enter radius of the circle: "))

for i in range((2 * radius) + 1):
    for j in range((2 * radius) + 1):
        # Calculate distance from the center
        dist = math.sqrt((i - radius) ** 2 + (j - radius) ** 2)
        # Print star if distance is close to the radius
        if radius - 0.5 < dist < radius + 0.5:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

LCM Least Common Multiple

```python
# Program to find the Least Common Multiple (LCM) of two numbers

def lcm(a, b):
    if a > b:
        greater = a
    else:
        greater = b

    while True:
        if (greater % a == 0) and (greater % b == 0):
            return greater
        greater += 1

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

print("The LCM of", num1, "and", num2, "is", lcm(num1, num2))
```

HCF Highest Common factor

```python
# Program to find the Highest Common Factor (HCF) of two numbers

def hcf(a, b):
    while b:
        a, b = b, a % b
    return a

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

print("The HCF of", num1, "and", num2, "is", hcf(num1, num2))
```


Compare performance of list vs. tuple for large datasets.


```python
# Compare performance of list vs. tuple for large datasets
import time

size = 10_000_000

# Create list and tuple
lst = list(range(size))
tpl = tuple(range(size))

# Measure access time for list
start = time.time()
_ = lst[5000000]
list_access_time = time.time() - start

# Measure access time for tuple
start = time.time()
_ = tpl[5000000]
tuple_access_time = time.time() - start

# Measure iteration time for list
start = time.time()
for i in lst:
    pass
list_iter_time = time.time() - start

# Measure iteration time for tuple
start = time.time()
for i in tpl:
    pass
tuple_iter_time = time.time() - start

print(f"List access time: {list_access_time:.6f} seconds")
print(f"Tuple access time: {tuple_access_time:.6f} seconds")
print(f"List iteration time: {list_iter_time:.6f} seconds")
print(f"Tuple iteration time: {tuple_iter_time:.6f} seconds")
```



Rotate elements of a tuple (e.g., shift left or right).

```python

def rotate_tuple_left(t, k=1):
    """Rotate tuple left by k positions."""
    k = k % len(t)
    return t[k:] + t[:k]

def rotate_tuple_right(t, k=1):
    """Rotate tuple right by k positions."""
    k = k % len(t)
    return t[-k:] + t[:-k]

# Example usage:
t = (1, 2, 3, 4, 5, 6, 7, 8)
print("Original tuple:", t)
print("Left rotated by 2:", rotate_tuple_left(t, 2))
print("Right rotated by 3:", rotate_tuple_right(t, 3))
```

Difference betwwe Tuple and List


Here is a table showing the **difference between List and Tuple in Python** in 10 key points:

| **Point**                | **List**                                          | **Tuple**                                         |
| ------------------------ | ------------------------------------------------- | ------------------------------------------------- |
| **1. Mutability**        | Mutable (can be changed after creation)           | Immutable (cannot be changed after creation)      |
| **2. Syntax**            | Defined using square brackets `[]`                | Defined using parentheses `()`                    |
| **3. Performance**       | Slower than tuple due to mutability               | Faster than list due to immutability              |
| **4. Use Case**          | Suitable for collections that may change          | Suitable for fixed data                           |
| **5. Methods Available** | Many built-in methods like `append()`, `remove()` | Fewer methods (`count()`, `index()`)              |
| **6. Memory Usage**      | Uses more memory                                  | Uses less memory                                  |
| **7. Iteration**         | Slower iteration                                  | Faster iteration                                  |
| **8. Nesting**           | Can be nested with other mutable objects          | Often used as keys in dictionaries (if hashable)  |
| **9. Hashable**          | Not hashable (can’t be dict keys)                 | Hashable if all elements are immutable            |
| **10. Error Prevention** | More error-prone due to possible changes          | Safer for fixed data, prevents accidental changes |

Given a string, count consonants (non-vowel letters, ignoring case). Example: "Hello World" → 7 (H, l, l, W, r, l, d).

```python

s = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0
for ch in s:
    if ch.isalpha() and ch not in vowels:
        count += 1
print("Number of consonants:", count)
```

```python

str1 = input("Enter first string: ").replace(" ", "").lower()
str2 = input("Enter second string: ").replace(" ", "").lower()

if sorted(str1) == sorted(str2):
    print("Anagram")
else:
    print("Not Anagram")
```

Compress a string by replacing repeated characters with the character and count. Example: "aaabbc" → "a3b2c1". If compressed string isn’t shorter, return original.

```python
s = input("Enter a string: ")
compressed = ""
count = 1

for i in range(1, len(s) + 1):
    if i < len(s) and s[i] == s[i - 1]:
        count += 1
    else:
        compressed += s[i - 1] + str(count)
        count = 1

if len(compressed) < len(s):
    print("Compressed string:", compressed)
else:
    print("Compressed string:", s)
```

6.
Find the longest substring in a string that is a palindrome (case-sensitive). Example: "babad" → "bab" (or "aba"), "cbbd" → "bb". If multiple exist, return the first.

```python
s = input("Enter a string: ")
max_palindrome = ""

for i in range(len(s)):
    for j in range(i + 1, len(s) + 1):
        substr = s[i:j]
        if substr == substr[::-1] and len(substr) > len(max_palindrome):
            max_palindrome = substr

if max_palindrome:
    print("Longest palindromic substring:", max_palindrome)
else:
    print("No palindrome found.")
```

7.
Given a string containing only '(', ')', '{', '}', '[', ']', check if the sequence is valid (properly nested and closed). Example: "({[]})" → "Valid", "(]" → "Invalid".

```python
s = input("Enter a string of brackets: ")
stack = []
pairs = {')': '(', '}': '{', ']': '['}
valid = True

for ch in s:
    if ch in '({[':
        stack.append(ch)
    elif ch in ')}]':
        if not stack or stack[-1] != pairs[ch]:
            valid = False
            break
        stack.pop()

if valid and not stack:
    print("Valid")
else:
    print("Invalid")
```

8.
Given a start word, end word, and a list of valid words, find the minimum number of steps to transform the start word to the end word by changing one letter at a time, where each intermediate word must be in the list. All words are same length, lowercase. Example: start="hit", end="cog", wordList=["hot","dot","dog","lot","log","cog"] → 5 (hit → hot → dot → dog → cog). Return 0 if impossible.

```python
from collections import deque

def ladder_length(beginWord, endWord, wordList):
    wordSet = set(wordList)
    if endWord not in wordSet:
        return 0
    queue = deque([(beginWord, 1)])
    while queue:
        word, steps = queue.popleft()
        if word == endWord:
            return steps
        for i in range(len(word)):
            for c in 'abcdefghijklmnopqrstuvwxyz':
                next_word = word[:i] + c + word[i+1:]
                if next_word in wordSet:
                    queue.append((next_word, steps + 1))
                    wordSet.remove(next_word)
    return 0

start = input("Enter start word: ")
end = input("Enter end word: ")
wordList = input("Enter word list (comma separated): ").split(',')

print("Minimum steps:", ladder_length(start, end, wordList))
```
