### Introduction to Matplotlib

**Matplotlib** is a powerful and widely used **data visualization library** in Python. It helps in creating static, animated, and interactive plots with ease. The most commonly used module in this library is `pyplot`, which mimics MATLAB’s plotting interface.

---

### 🔧 Basic Features:

| Feature       | Description                                        |
| ------------- | -------------------------------------------------- |
| 2D Plotting   | Line graphs, bar charts, histograms, scatter, etc. |
| Customization | Titles, labels, legends, colors, styles            |
| Exporting     | Save plots as images (PNG, PDF, SVG, etc.)         |
| Integration   | Works with NumPy, pandas, Jupyter, etc.            |

---

### ✍️ Basic Example:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)          # Line plot
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

---

### 📊 Types of Plots:

| Plot Type    | Function        |
| ------------ | --------------- |
| Line Plot    | `plt.plot()`    |
| Bar Chart    | `plt.bar()`     |
| Histogram    | `plt.hist()`    |
| Scatter Plot | `plt.scatter()` |
| Pie Chart    | `plt.pie()`     |




## 🧠 How It Works: Big Picture

Matplotlib has two main interfaces:

| Interface      | Description                                | Example                        |
|----------------|--------------------------------------------|--------------------------------|
| `pyplot`       | Simple, MATLAB-like interface (easy to use) | `plt.plot(...)`, `plt.show()` |
| `Object-Oriented` | Advanced control, figure & axes objects   | `fig, ax = plt.subplots()`    |

---

## ✨ Basic Syntax (Pyplot Style)

```python
import matplotlib.pyplot as plt

# X and Y values
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)               # Line plot
plt.title("My Line Graph")   # Title
plt.xlabel("X Axis")         # X label
plt.ylabel("Y Axis")         # Y label
plt.grid(True)               # Show grid
plt.show()                   # Display the plot
```

## 🔧 Customization Examples

```python
plt.plot(x, y, color='red', linestyle='--', marker='o', linewidth=2)
plt.legend(["My Data"])
```

| Customization     | Use                            |
|-------------------|--------------------------------|
| `color='red'`     | Line color                     |
| `linestyle='--'`  | Dashed line                    |
| `marker='o'`      | Dots on data points            |
| `linewidth=2`     | Thickness of line              |
| `legend([...])`   | Add legend                     |

---

## 🧱 Figure, Axes, and Subplots (Object-Oriented)

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Line Chart")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

| Component | Meaning                                       |
|-----------|-----------------------------------------------|
| `Figure`  | The entire canvas (can contain multiple plots)|
| `Axes`    | A single plot area                            |
| `Subplot` | Multiple plots in one figure                  |

---

## 📥 Saving Plots

```python
plt.savefig("myplot.png", dpi=300, bbox_inches='tight')
```

| Option            | Purpose                    |
|-------------------|----------------------------|
| `dpi=300`         | High-resolution image      |
| `bbox_inches='tight'` | Trim white space       |

Draw 6 plot


### ✅ Options for 6 subplots:

1. **2 rows × 3 columns** → `subplot(2, 3, i)` where `i` = 1 to 6
2. **3 rows × 2 columns** → `subplot(3, 2, i)` where `i` = 1 to 6
3. **6 rows × 1 column** → `subplot(6, 1, i)` for vertical stacking
4. **1 row × 6 columns** → `subplot(1, 6, i)` for horizontal layout

---

### ✅ Example (2 rows × 3 columns):

```python
import matplotlib.pyplot as plt

for i in range(1, 7):
    plt.subplot(2, 3, i)  # 2 rows, 3 columns, subplot i
    plt.plot([1, 2, 3], [j*i for j in [1, 2, 3]])
    plt.title(f"Plot {i}")

plt.tight_layout()
plt.show()
```

Here, the argument `(2, 3, i)` means:

* 2 rows
* 3 columns
* `i`th position (from 1 to 6)

Multiple Bar Plots

````python
import matplotlib.pyplot as plt
import numpy as np

barwidth = 0.25

IT = [12, 22, 3, 4, 5]
ECE = [2, 33, 23, 3, 15]
CSE = [2, 11, 23, 8, 9]

br1 = np.arange(len(IT))
br2 = [x + barwidth for x in br1]
br3 = [x + barwidth for x in br2]

plt.bar(br1, IT, color='r', width=barwidth, edgecolor='gray', label='IT')
plt.bar(br2, ECE, color='g', width=barwidth, edgecolor='gray', label='ECE')
plt.bar(br3, CSE, color='b', width=barwidth, edgecolor='gray', label='CSE')

plt.xlabel('Branch')
plt.ylabel('Students')
plt.title('Students in Different Branches')
plt.xticks([r + barwidth for r in range(len(IT))], ['2018', '2019', '2020', '2021', '2022'])
plt.legend()
plt.show()
````

# Scatter Plot

Plotting With Keyword String

````python
data = {
    'a': np.arange(50),
    'c': np.random.randint(0, 50, 50),
    'd': np.random.randn(50)
}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.title("Scatter Plot with Color and Size")
plt.show()
````

Here's the explanation step by step:

---

### 🧮 1. Create a `data` dictionary:

```python
data = {
    'a': np.arange(50),
    'c': np.random.randint(0, 50, 50),
    'd': np.random.randn(50)
}
```

| Key   | Meaning                                                        |
| ----- | -------------------------------------------------------------- |
| `'a'` | Numbers from 0 to 49 (X-axis values)                           |
| `'c'` | 50 random integers from 0 to 49 (used for **color**)           |
| `'d'` | 50 random numbers from normal distribution (used for **size**) |

---

### 🧮 2. Add a new column `'b'`:

```python
data['b'] = data['a'] + 10 * np.random.randn(50)
```

✅ `'b'` is created by adding noise (random variation) to `'a'`. This makes the Y-values for scatter points.

---

### 🧮 3. Modify `'d'` to make it usable for size:

```python
data['d'] = np.abs(data['d']) * 100
```

✅ Takes the absolute value (no negative sizes), then multiplies by 100 to scale the point sizes.

---

### 🎨 4. Plot the scatter plot:

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
```

| Parameter   | Meaning                                   |
| ----------- | ----------------------------------------- |
| `'a'`       | X-axis values                             |
| `'b'`       | Y-axis values                             |
| `c='c'`     | Color of each point based on `'c'`        |
| `s='d'`     | Size of each point based on `'d'`         |
| `data=data` | Tells `matplotlib` to use the `data` dict |

---

### 🏷️ 5. Labels and title:

```python
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.title("Scatter Plot with Color and Size")
```

Adds axis labels and a title to the plot.

---

### 📊 6. Show the plot:

```python
plt.show()
```

---

This will give you a colorful, varied-size **scatter plot** where:

* X = `'a'`
* Y = `'b'`
* Point color = `'c'`
* Point size = `'d'`

# 3D Contour Plot with Matplotlib

This section demonstrates how to create 3D contour plots using Matplotlib in Python.

## 1. Import Required Libraries

We'll need `matplotlib`, `numpy`, and `mpl_toolkits.mplot3d` for 3D plotting.

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

## 2. Generate 3D Data for Contour Plot

Let's create a meshgrid for X and Y, and compute Z as a function of X and Y.

````python
# Generate grid data
x = np.linspace(-4, 4, 50)
y = np.linspace(-4, 4, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
````

## 3. Create 3D Contour Plot using Matplotlib

We'll use `Axes3D` and `ax.contour3D` to plot the 3D contour.

````python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot 3D contour
contour = ax.contour3D(X, Y, Z, 50, cmap='viridis')

plt.show()
````

## 4. Customize 3D Contour Plot (Labels, Colors, etc.)

Add axis labels, a title, and a color bar for better visualization.

````python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

contour = ax.contour3D(X, Y, Z, 50, cmap='plasma')
ax.set_title("3D Contour Plot Example")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

# Add color bar
fig.colorbar(contour, ax=ax, shrink=0.6, aspect=10, label='Z value')

plt.show()
````
