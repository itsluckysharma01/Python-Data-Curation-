### Introduction to Matplotlib

**Matplotlib** is a powerful and widely used **data visualization library** in Python. It helps in creating static, animated, and interactive plots with ease. The most commonly used module in this library is `pyplot`, which mimics MATLAB‚Äôs plotting interface.

---

### üîß Basic Features:

| Feature       | Description                                        |
| ------------- | -------------------------------------------------- |
| 2D Plotting   | Line graphs, bar charts, histograms, scatter, etc. |
| Customization | Titles, labels, legends, colors, styles            |
| Exporting     | Save plots as images (PNG, PDF, SVG, etc.)         |
| Integration   | Works with NumPy, pandas, Jupyter, etc.            |


````python
import numpy as np

#evenly sampled time as 200ms intervals
t=np.arange(0. , 5. , 0.2)

#red dashed, blue square, green triangle
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.title("Multiple Line Styles")
plt.show()
````

| Data          | Style Code | What It Plots              |
| ------------- | ---------- | -------------------------- |
| `t` vs `t`    | `'r--'`    | Red dashed line (`y = t`)  |
| `t` vs `t**2` | `'bs'`     | Blue squares (`y = t¬≤`)    |
| `t` vs `t**3` | `'g^'`     | Green triangles (`y = t¬≥`) |


# Scatter Plot

Plotting With Keyword String

````python
data = {
    'a': np.arange(50),
    'c': np.random.randint(0, 50, 50),
    'd': np.random.randn(50)
}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.title("Scatter Plot with Color and Size")
plt.show()
````

Here's the explanation step by step:

---

### üßÆ 1. Create a `data` dictionary:

```python
data = {
    'a': np.arange(50),
    'c': np.random.randint(0, 50, 50),
    'd': np.random.randn(50)
}
```

| Key   | Meaning                                                        |
| ----- | -------------------------------------------------------------- |
| `'a'` | Numbers from 0 to 49 (X-axis values)                           |
| `'c'` | 50 random integers from 0 to 49 (used for **color**)           |
| `'d'` | 50 random numbers from normal distribution (used for **size**) |

---

### üßÆ 2. Add a new column `'b'`:

```python
data['b'] = data['a'] + 10 * np.random.randn(50)
```

‚úÖ `'b'` is created by adding noise (random variation) to `'a'`. This makes the Y-values for scatter points.

---

### üßÆ 3. Modify `'d'` to make it usable for size:

```python
data['d'] = np.abs(data['d']) * 100
```

‚úÖ Takes the absolute value (no negative sizes), then multiplies by 100 to scale the point sizes.

---

### üé® 4. Plot the scatter plot:

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
```

| Parameter   | Meaning                                   |
| ----------- | ----------------------------------------- |
| `'a'`       | X-axis values                             |
| `'b'`       | Y-axis values                             |
| `c='c'`     | Color of each point based on `'c'`        |
| `s='d'`     | Size of each point based on `'d'`         |
| `data=data` | Tells `matplotlib` to use the `data` dict |

---

### üè∑Ô∏è 5. Labels and title:

```python
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.title("Scatter Plot with Color and Size")
```

Adds axis labels and a title to the plot.

---

### üìä 6. Show the plot:

```python
plt.show()
```

---

This will give you a colorful, varied-size **scatter plot** where:

* X = `'a'`
* Y = `'b'`
* Point color = `'c'`
* Point size = `'d'`



# Categorical Variables

````python
names=['group_a', 'group_b', 'group_c']
values=[1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle("CATEGORICAL PLOTTING")
plt.show()
````

Multiple Figure and Axis

````python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')


````
